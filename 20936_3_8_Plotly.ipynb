```python
df = pd.read_csv('Wisconsin_Diagnostic_Breast_Cancer_(WDBC).csv')
df
```

```python
classes = df['diagnosis']                
features = df.drop(['Unnamed: 32','id','diagnosis'], axis = 1 )
features.head()
```

```python
data = pd.concat([classes, features.iloc[:,0]], axis=1)
data = pd.melt(data, id_vars="diagnosis",
                     var_name="features",
                     value_name='value')
data
```

```python
fig = px.violin(data_frame=df,
                x="diagnosis",
                y="radius_mean",
                color="diagnosis",
                box=True)
fig.show()
```

```python
classes = df['diagnosis']                
features = df.drop(['Unnamed: 32','id','diagnosis'], axis = 1 )
features = ((features - features.mean()) / features.std())
data = features
data.head()
```

```python
data = pd.concat([classes, features.iloc[:,0:6]], axis=1)
data = pd.melt(data, id_vars="diagnosis",
                     var_name="features",
                     value_name='value')
fig = px.violin(data_frame=data,
                x="features",
                y="value",
                color="diagnosis",
                box=True)
fig.show()
```

```python
data = pd.concat([classes, features.iloc[:,0:4]], axis=1)
fig = px.scatter_matrix(data_frame=data,
                        dimensions=data.columns[1:],
                        color="diagnosis")
fig.update_traces(diagonal_visible=False)
fig.show()
```

```python
fig = px.imshow(features.corr(), text_auto=True)
fig.data[0].text=features.corr().round(decimals=2)
fig.data[0].texttemplate="%{text}"
fig.show()
```

##PCA

```python
df = pd.read_csv("iris_data_set.csv")
df = df.drop(df.index[-1])
classes = df.iloc[:, -1]
features = df.iloc[:, 0:4]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

features_norm = StandardScaler().fit_transform(features)

pca = PCA(n_components = 3)
principalComponents = pca.fit_transform(features_norm)
principalDf = pd.DataFrame(data = principalComponents,
                           columns = ['PC 1', 'PC 2', 'PC 3'])
principalDf.head()
```

```python
df_pca = pd.concat([principalDf, classes], axis = 1)
fig = px.scatter_3d(data_frame=df_pca,
                    x="PC 1", y="PC 2", z="PC 3",
                    color="class")
fig.show()
```
